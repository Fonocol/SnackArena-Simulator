```python
import os
import numpy as np
import matplotlib.pyplot as plt
import random

```

```python
channel_names = [
    "Agent", "target", "wall", "facing"
]

def show_minimap(tensor):
    num_channels = tensor.shape[0]

    fig, axes = plt.subplots(1, num_channels + 1, figsize=(15, 4))

    # Afficher chaque canal séparément
    for i in range(num_channels):
        axes[i].imshow(tensor[i], cmap="gray")
        axes[i].set_title(channel_names[i])
        axes[i].axis("off")

    # Fusion RGB : agent = rouge, cible = vert, mur = bleu
    combined = np.zeros((tensor.shape[1], tensor.shape[2], 3), dtype=np.float32)
    combined[..., 0] = tensor[0]  # rouge : agent
    combined[..., 1] = tensor[1]  # vert  : target
    combined[..., 2] = tensor[2] + 0.5 * tensor[3]  # bleu : murs + facing (en demi-intensité)

    axes[-1].imshow(combined)
    axes[-1].set_title("Fusion RGB")
    axes[-1].axis("off")

    plt.tight_layout()
    plt.show()
```

```python
tensor_dir = r"C:\Users\etudiant\Documents\MesEtudes\projets\snack\data\minimap_logs"
tensor_files = [f for f in os.listdir(tensor_dir) if f.endswith(".npy")]

for step in tensor_files:
    print("affichage de :", step)
    file_to_load = os.path.join(tensor_dir,step)
    tensor = np.load(file_to_load)
    show_minimap(tensor)

```
